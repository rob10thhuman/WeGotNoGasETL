```python
import pandas as pd
from sqlalchemy import create_engine
```

### Extract CSV from Kaggle.com into a DataFrame

```python
#create df for kaggle dataset 'ElectricCarData_Clean.csv'
ev_csv_file = "Resources/ElectricCarData_Clean.csv"
ev_csv_df = pd.read_csv(ev_csv_file)
ev_csv_df.head()
```

### Transform ev_csv DataFrame

```python
#CSV: convert km to miles on cols: Top_Speed, Range_Km
top_speed = []
for x in ev_csv_df['TopSpeed_KmH']:
    mph = round(ev_csv_df['TopSpeed_KmH'] * 0.6214, 0)
    top_speed.append(mph)
    
top_speed = pd.DataFrame(top_speed[1])
ev_csv_df['Top_speed_mph'] = top_speed

range_df = []
for y in ev_csv_df['Range_Km']:
    mph2 = round(ev_csv_df['Range_Km'] * 0.6214, 0)
    range_df.append(mph2)
    
range_df = pd.DataFrame(range_df[1])
ev_csv_df['Range_miles'] = range_df
ev_csv_df.head()
```

```python
# Create a filtered dataframe from specific columns
ev_csv_cols = ["Brand", "Model", "PowerTrain", 'BodyStyle', 'Seats', "PlugType", "Top_speed_mph", "Range_miles", "AccelSec"]
ev_csv_transformed= ev_csv_df[ev_csv_cols].copy()

# Rename the column headers
ev_csv_transformed = ev_csv_transformed.rename(columns={"Brand": "Make",
                                                          "Top_speed_mph": "TopSpeed_mph"})
ev_csv_transformed.head()
```

```python
ev_csv_transformed.to_csv(r'ev_csv_transformed.csv', index = False)
```
